```python
struct_df['distance'] = struct_df['endpoints'].apply(lambda x: np.linalg.norm(np.array(x[0]) - np.array(x[1])))

trimmed_df = struct_df.loc[struct_df.iloc[:,1].apply(len) != 1]

trimmed_df = trimmed_df.loc[~(abs(trimmed_df['time']) > 10)]
```
